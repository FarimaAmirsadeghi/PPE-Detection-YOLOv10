```python
import os
import shutil
from collections import defaultdict
import random

# Paths
label_dir = "/content/drive/MyDrive/dataset2/labels/all"
image_dir = "/content/drive/MyDrive/dataset2/images/all"
output_base = "/content/drive/MyDrive/dataset2"

# Output dirs
for split in ['train', 'val']:
    os.makedirs(f"{output_base}/images/{split}", exist_ok=True)
    os.makedirs(f"{output_base}/labels/{split}", exist_ok=True)

# Group label files by class
class_to_files = defaultdict(set)
all_files = set()

for file in os.listdir(label_dir):
    if not file.endswith(".txt"):
        continue
    path = os.path.join(label_dir, file)
    with open(path, "r") as f:
        lines = f.readlines()
    fname = file[:-4]  # without .txt
    all_files.add(fname)
    for line in lines:
        if line.strip():
            class_id = int(line.split()[0])
            class_to_files[class_id].add(fname)

# Collect validation set (10% per class)
val_set = set()
for cid, files in class_to_files.items():
    n_val = max(1, int(0.1 * len(files)))
    val_samples = random.sample(list(files), n_val)
    val_set.update(val_samples)

# Final split
for fname in all_files:
    split = 'val' if fname in val_set else 'train'
    # Copy label
    shutil.copy(f"{label_dir}/{fname}.txt", f"{output_base}/labels/{split}/{fname}.txt")
    # Copy image
    for ext in [".jpg", ".png"]:
        img_path = f"{image_dir}/{fname}{ext}"
        if os.path.exists(img_path):
            shutil.copy(img_path, f"{output_base}/images/{split}/{fname}{ext}")
            break  # found correct image

# Summary
print(f"Stratified split complete.")
print(f"Validation: {len(val_set)} files (includes all classes)")
print(f"Training: {len(all_files) - len(val_set)} files")
```

metrics for original dataset

```python
import matplotlib.pyplot as plt
from collections import Counter
import os
import yaml

# Path  data.yaml
yaml_path = "/content/drive/MyDrive/dataset2/data.yaml"
filtered_labels_path = "/content/drive/MyDrive/dataset2/labels"
splits = ["train", "val", "test"]

# Read names from data.yaml
with open(yaml_path, 'r') as stream:
    data_yaml = yaml.safe_load(stream)

class_names = data_yaml['names']
class_counts = Counter() # Define class_counts as a Counter object

for split in splits:
    label_dir = os.path.join(filtered_labels_path, split)
    if not os.path.exists(label_dir):
        print(f" Not exist {label_dir}, omitted.")
        continue

    label_files = [f for f in os.listdir(label_dir) if f.endswith(".txt")]
    for file in label_files:
        with open(os.path.join(label_dir, file), "r") as f:
            lines = f.readlines()
        for line in lines:
            class_id = int(line.strip().split()[0])
            class_counts[class_id] += 1


print("\nFinal Distribution")
for class_id, count in class_counts.items():
    if class_id < len(class_names):
        print(f"{class_names[class_id]}: {count} examples")
    else:
        print(f"Clase ID {class_id}: {count} examples")

# Graph
plt.figure(figsize=(12, 6))

# check  class_id
labels = [class_names[i] if i < len(class_names) else f"ID {i}" for i in class_counts.keys()]
values = list(class_counts.values())

plt.bar(labels, values, color='skyblue', edgecolor='black')
plt.xlabel("Categories", fontsize=12)
plt.ylabel("Number of instances", fontsize=12)
plt.title("Categories Distribution", fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
from ultralytics import YOLO
import torch

torch.cuda.empty_cache()
#  training
model = YOLO("yolov10m.pt")

model.train(
    data="/content/drive/MyDrive/dataset2/data.yaml",
    project="/content/filtered_runs",
    name="train_resume",
    epochs=100,                # <--- training 100 epochs
    batch=16,
    workers=2,
    imgsz=640,
    optimizer='AdamW',
    patience=20,
    lr0=1e-4,
    save=True,
    save_period=1,           # save each epoch
)
```
